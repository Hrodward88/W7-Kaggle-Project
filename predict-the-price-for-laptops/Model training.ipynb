```python
def convert_RAM_to_numeric(strings): #Convert RAM to numeric
    result = []
    for string in strings:
        if len(string) >= 2 and string[-2] == 'G':
            string = string[:-2]
            result.append(pd.to_numeric(string))
        else:
            result.append(pd.to_numeric(string))
    return result

strings= ['24GN', '8', '12.4', '5900']

convert_string_to_numeric(strings)
```

```python
training = training.rename(columns={'RAM': 'RAM (GB)', 'Screen Size' : 'Screen Size (In.)', })
training.head()
```

```python
training['RAM (GB)'] = convert_string_to_numeric(training['RAM (GB)'])
```

```python
training.head()
```

```python
plt.figure(figsize=(15, 10))

sns.set(style='white')

mask=np.triu(np.ones_like(training.corr(), dtype=bool))

cmap=sns.diverging_palette(0, 10, as_cmap=True)


sns.heatmap(training.corr(),
           mask=mask,
          cmap=cmap,
          center=0,
          square=True,
          annot=True,
          linewidths=0.5,
          cbar_kws={'shrink': 0.5});
```
